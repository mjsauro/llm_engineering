# Mafia Game

## Social Deduction Game with AI Players

This is an interactive Mafia game where 1 moderator AI controls the game flow and 3 player AIs compete with hidden roles. Players must work together or against each other to achieve their team's win condition.

### How to Play:
1. Install required models (if needed)
2. Configure the game settings
3. Initialize the game (roles will be assigned automatically)
4. Execute night phases (Mafia kill, Doctor protect, Detective investigate)
5. Execute day phases (discussion and voting)
6. Game continues until a win condition is met

### Roles:
- **Mafia**: Eliminate Villagers until Mafia equals or outnumbers them
- **Doctor**: Protect players from Mafia kills
- **Detective**: Investigate players to learn if they are Mafia
- **Villager**: Use deduction to identify and eliminate Mafia

### Win Conditions:
- **Villagers win**: All Mafia eliminated
- **Mafia win**: Mafia count >= Villager count

### Learning Objectives:
- Understand multi-AI interaction with hidden information
- See how AI models can play social deduction games
- Learn about private vs public communication channels
- Experience role-based gameplay with conversation history tracking

## Night Phase

Execute the night phase. During the night:
1. Mafia coordinates privately and chooses a kill target
2. Doctor chooses a player to protect
3. Detective chooses a player to investigate
4. Actions are resolved (kill succeeds unless protected)
5. Eliminated player is revealed

**Run this cell to execute a night phase.**


```python
# Execute night phase
if game.game_over:
    print("=" * 80)
    print("GAME OVER!")
    print("=" * 80)
    print()
    print(f"Winner: {game.winner}")
    print()
    print("Game has ended. To start a new game, reinitialize the game instance.")
else:
    night_result = game.execute_night_phase()
    
    # Check win condition
    winner = game.check_win_condition()
    if winner:
        game.game_over = True
        game.winner = winner
        print("=" * 80)
        print("ðŸ† GAME OVER!")
        print("=" * 80)
        print(f"Winner: {winner}")
        print("=" * 80)
    else:
        print("\nNight phase complete. Proceed to day phase.")
```

## Day Phase

Execute the day phase. During the day:
1. Moderator announces the day and any eliminations
2. Public discussion: All living players can speak, accuse, defend
3. Voting: Players vote to eliminate a suspect
4. Player with majority votes is eliminated and their role is revealed
5. Win condition is checked

**Run this cell to execute a day phase.**


```python
# Execute day phase
if game.game_over:
    print("=" * 80)
    print("GAME OVER!")
    print("=" * 80)
    print()
    print(f"Winner: {game.winner}")
    print()
    print("Game has ended. To start a new game, reinitialize the game instance.")
else:
    day_result = game.execute_day_phase()
    
    # Check win condition
    winner = game.check_win_condition()
    if winner:
        game.game_over = True
        game.winner = winner
        print("=" * 80)
        print("ðŸ† GAME OVER!")
        print("=" * 80)
        print(f"Winner: {winner}")
        print("=" * 80)
    else:
        print("\nDay phase complete. Proceed to next night phase.")
```

## How It Works

### Game Mechanics
- **Moderator**: Controls game flow, narrates events, doesn't play
- **Role Assignment**: Randomly assigns 1 Mafia, 1 Doctor, 1 Detective to 3 players
- **Mafia Knowledge**: Mafia players know each other's identities
- **Other Players**: Don't know anyone's role initially

### Night Phase
- **Mafia**: Coordinates privately and chooses a kill target
- **Doctor**: Chooses one player to protect (prevents kill if target matches)
- **Detective**: Chooses one player to investigate (learns if they are Mafia/Villager)
- **Resolution**: Kill succeeds unless protected by Doctor

### Day Phase
- **Discussion**: All living players can speak publicly, accuse, defend, share information
- **Voting**: Players vote to eliminate a suspect
- **Elimination**: Player with majority votes is eliminated, role revealed

### Win Conditions
- **Villagers Win**: All Mafia eliminated
- **Mafia Win**: Mafia count >= Villager count
- Checked after each elimination (night or day)

### Communication System
- **Public Channel**: All players see day discussion
- **Private Channel**: Mafia can coordinate privately at night
- **Conversation History**: Full history maintained for each AI (following day1.ipynb pattern)
- **System + User Prompts**: Each AI gets role-specific system prompt and conversation history in user prompt
